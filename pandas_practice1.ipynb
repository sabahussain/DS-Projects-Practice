```python
import numpy as np
import pandas as pd
```

# Load the Titanic dataset into a DataFrame. Display the first 5 rows.

```python
df=pd.read_csv('train.csv')
df.head()
```

# How many rows and columns are there?

```python
df.shape
```

# What are the column names and their data types?

```python
df.columns
```

```python
df.dtypes
```

# Show the number of missing values in each column.

```python
df.isnull().sum()
```

#  What is the average age of passengers?

```python
df.Age.mean()
```

# Select all passengers who were female.

```python
df.loc[df.Sex=='female']
```

# Find passengers older than 50.

```python
df.loc[df.Age>50]
```

# Get the names of passengers who paid a fare greater than 100.

```python
df.loc[df.Fare>100, "Name"]
```

# Find passengers who embarked at ‚ÄúS‚Äù and survived.

```python
df.loc[(df.Embarked=='S') & (df.Survived==1)]
```

# Show all children (age < 12) who were in 3rd class.

```python
df.loc[(df.Age<12) & (df.Pclass==3)]
```

# What is the survival rate (percentage) overall?

```python
totalpassengers= len(df.PassengerId)
survived= df.Survived.sum()
perc= (survived/totalpassengers)* 100
print(perc)
# perc = df['Survived'].mean() * 100
# print(perc)
```

# What is the survival rate by gender?

```python
df.groupby('Sex').Survived.mean()*100
```

# What is the average age of survivors vs non-survivors?

```python
df.groupby('Survived').Age.mean()
```

# Which passenger class (Pclass) had the highest survival rate?

```python
# df.loc[df.Survived==1].groupby(['Pclass']).Survived.count().idxmax()

survival_rate = df.groupby('Pclass')['Survived'].mean()
print(survival_rate)

highest_rate_class = survival_rate.idxmax()
print("Class with highest survival rate:", highest_rate_class)
```

# What was the average fare paid per class?

```python
df.groupby('Pclass').Fare.mean()
```

# Who paid the highest fare? Display their details.

```python
idx= df.Fare.idxmax()
```

```python
idx
```

```python
df.loc[idx]
```

# Sort passengers by age (youngest to oldest). Show top 10 oldest.

```python
df.dropna(subset=['Age']).sort_values(by='Age').tail(10)
```

# Find the 5 passengers with the most expensive tickets who did not survive.

```python
df.loc[df.Survived==0].sort_values(by='Fare', ascending=False).head()
```

# Compare the survival rate of children (<12) vs adults (>=12).

```python
"""children= df.loc[(df.Survived==1) & (df.Age<12)]
adults= df.loc[(df.Survived==1) & (df.Age>=12)]
print(len(children),"children survived out of ", sum(df.Age<12))
print(len(adults),"adults survived out of ", sum(df.Age>=12))"""
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 12 else 'Adult')
survival_rates = df.groupby('AgeGroup')['Survived'].mean()
print(survival_rates)
```

# For each embarkation port, find the average fare and survival rate.

```python
df.groupby('Embarked').agg({"Fare":'mean', "Survived": 'mean'})
```
