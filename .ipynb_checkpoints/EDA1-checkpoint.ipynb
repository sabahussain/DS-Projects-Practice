### What is EDA (Exploratory Data Analysis)?

Exploratory Data Analysis (EDA) is the process of analyzing a dataset to:

Understand its structure (columns, data types, missing values)

Summarize main characteristics (statistics, distributions, outliers)

Find relationships between variables (correlations, patterns)

Visualize trends (graphs, plots, charts)

In simple words: EDA = ‚Äúgetting to know your data‚Äù before applying machine learning or drawing conclusions.

### Typical EDA Steps

**Data Inspection**

Look at first/last few rows (df.head(), df.info(), df.describe())

Check data types, missing values, duplicates

**Univariate Analysis (single variable)**

Distribution of numeric features (histograms, boxplots)

Frequency of categorical features (bar plots)

**Bivariate/Multivariate Analysis (relationships)**

Correlation between numerical features

Survival rate by class/gender (like Titanic)

Scatter plots, groupby summaries

**Handling Missing & Outliers**

Fill, drop, or impute missing values

Detect outliers (boxplot, IQR)

**Feature Engineering (optional)**

Create new meaningful columns

### What is Correlation?

Correlation tells us how strongly two variables move together.

If one variable goes up when the other goes up, that‚Äôs positive correlation.

If one goes up when the other goes down, that‚Äôs negative correlation.

If they don‚Äôt affect each other, that‚Äôs no correlation.

### The Range of Correlation

Correlation values always lie between -1 and +1:

+1 ‚Üí Perfect Positive Correlation
Example: Temperature in ¬∞C and Temperature in ¬∞F. (As ¬∞C increases, ¬∞F always increases in a fixed way.)

-1 ‚Üí Perfect Negative Correlation
Example: Speed of a car and time taken to reach a destination (higher speed ‚Üí less time).

0 ‚Üí No Correlation
Example: Shoe size and IQ (completely unrelated).


# 26. Compare the success of movies before and after the year 2000 (rating + revenue).

```python
df['Era'] = df['Year'].apply(lambda x: 'Before 2000' if x < 2000 else '2000 & After')
comparison = df.groupby("Era")[['Rating', 'Revenue']].mean()
print(comparison)
```

```python
comparison = df.groupby(df['Year'] < 2000)[['Rating', 'Revenue']].mean()
print(comparison)
```

```python
comparison.index=comparison.index.map({True: "Before 2000", False: "2000 & After"})
```

```python
comparison
```

# 27. What is the survival rate of movies with high revenue? (e.g., % of movies earning above $100M).

```python
high_revenue_movies = df['Revenue'] > 100
high_revenue_movies.mean() * 100
```

# 28. Are movies with multiple genres rated differently than single-genre movies?

```python
df['MultiGenre'] = df['Genre'].str.contains(',')
rating_comparison = df.groupby('MultiGenre')['Rating'].mean()
print(rating_comparison)
```

# 29. Create a bar chart showing the top 10 highest-grossing movies.

```python
top10 = df.sort_values(by="Revenue", ascending=False).head(10)
plt.bar(top10["Title"], top10["Revenue"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Movie Title")
plt.ylabel("Revenue (in $M)")
plt.title("Top 10 Highest-Grossing Movies")
plt.show()
```

# 30. Which 3 movies earned the highest revenue?

```python
# 33. Create a new column Decade (e.g., 1990s, 2000s, 2010s). Which decade produced the highest-rated movies?
```

```python
df['Decade'] = (df['Year'] // 10) * 10
df['Decade'] = df['Decade'].astype(str) + "s"
```

```python
df.groupby('Decade').Rating.mean().sort_values(ascending=False).idxmax()
```
