
# 12. Do longer movies get better/worse ratings?


```python
movies = df[df['type'] == 'Movie'].copy()

# Convert duration to numeric
movies['Runtime'] = movies['duration'].str.replace(" min", "").astype(float)

# Average runtime by parental rating
runtime_by_rating = movies.groupby('rating')['Runtime'].mean().sort_values(ascending=False)
print(runtime_by_rating)

plt.figure(figsize=(10,6))
runtime_by_rating.plot(kind='bar', color='skyblue')
plt.title("Average Runtime by Parental Rating")
plt.xlabel("Rating")
plt.ylabel("Average Runtime (minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
bins = [0, 90, 120, 300]
labels = ['Short (<90)', 'Medium (90-120)', 'Long (>120)']
movies['RuntimeGroup'] = pd.cut(movies['Runtime'], bins=bins, labels=labels, right=False)

runtime_vs_rating = pd.crosstab(movies['RuntimeGroup'], movies['rating'])
print(runtime_vs_rating)
```

# 14. Are there recurring themes in descriptions (e.g., "love", "crime", "friendship")?

```python
themes = ['love', 'crime', 'friendship', 'family', 'war', 'murder', 'life']
```

```python
theme_counts = {}

for theme in themes:
    theme_counts[theme] = df['description'].str.contains(
        theme, case=False, na=False
    ).sum()
```

```python
theme_counts
```

```python
theme_series = pd.Series(theme_counts).sort_values(ascending=False)
print(theme_series)
```

```python
plt.figure(figsize=(8,5))
theme_series.plot(kind='bar', color='purple')
plt.title("Recurring Themes in Netflix Descriptions")
plt.xlabel("Theme")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

# 15. Can you identify clusters of shows based on keywords in their descriptions?

```python
themes = {
    'Crime': ['crime', 'murder', 'police', 'detective', 'investigation'],
    'Romance': ['love', 'relationship', 'romance', 'marriage'],
    'Comedy': ['comedy', 'funny', 'humor', 'laugh'],
    'Sci-Fi': ['future', 'space', 'alien', 'technology'],
    'War': ['war', 'army', 'soldier', 'battle', 'history']
}
```

```python
themes
```

```python
def assign_theme(description):
    if pd.isna(description):
        return 'Other'
    desc = description.lower()
    for theme, keywords in themes.items():
        if any(word in desc for word in keywords):
            return theme
    return 'Other'
```

```python
df['Theme'] = df['description'].apply(assign_theme)
```

```python
df['Theme'].value_counts()
```

```python
df['Theme'].value_counts().plot(kind='bar', figsize=(8,5), color='teal')
plt.title("Content Themes Based on Description Keywords")
plt.xlabel("Theme")
plt.ylabel("Number of Titles")
plt.show()
```

# 16. Compare content from USA vs India vs UK (most common genres, average release year).

```python
countries = ['United States', 'India', 'United Kingdom']

df_countries = df[df['country'].str.contains(
    'United States|India|United Kingdom',
    case=False,
    na=False
)].copy()
```

```python
df_genre = df_countries.copy()

df_genre['Country'] = df_genre['country'].str.split(', ')
df_genre['Genre'] = df_genre['listed_in'].str.split(', ')

df_genre = df_genre.explode('Country').explode('Genre')
```

```python
df_genre
```

```python
df_genre = df_genre[df_genre['Country'].isin(countries)]
```

```python
df_genre
```

```python
top_genres = (
    df_genre
    .groupby(['Country', 'Genre'])
    .size()
    .reset_index(name='Count')
)

top_genres_per_country = top_genres.loc[
    top_genres.groupby('Country')['Count'].idxmax()
]

print(top_genres_per_country)
```

```python
avg_release_year = (
    df_genre
    .groupby('Country')['release_year']
    .mean()
    .round()
)

print(avg_release_year)
```

# 17. Are Netflix Originals (if flagged) rated better than non-originals?

```python
df['IsNetflixOriginal'] = df['description'].str.contains(
    'Netflix Original',
    case=False,
    na=False
)
```

```python
rating_dist = (
    df.groupby(['IsNetflixOriginal', 'rating'])
      .size()
      .unstack(fill_value=0)
)
```

```python
rating_dist
```

```python
rating_percent = rating_dist.div(rating_dist.sum(axis=1), axis=0) * 100
rating_percent
```

```python
rating_percent.T.plot(kind='bar', figsize=(10,6))
plt.title("Parental Rating Distribution: Netflix Originals vs Non-Originals")
plt.xlabel("Rating")
plt.ylabel("Percentage of Titles")
plt.xticks(rotation=45)
plt.legend(title="Is Netflix Original")
plt.show()
```

```python
# Netflix Originals have a much higher proportion of TV-MA / TV-14
# Non-originals have more PG, TV-Y, family content
```

# 18.  Does Netflix add more international (non-USA) content over time?

```python
df['IsUSA'] = df['country'].str.contains(
    'United States',
    case=False,
    na=False
)
```

```python
df['YearAdded'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
```

```python
content_growth = (
    df.groupby(['YearAdded', 'IsUSA'])
      .size()
      .unstack(fill_value=0)
)

content_growth.columns = ['International', 'USA']
content_growth.head()
```

```python
plt.figure(figsize=(12,6))
content_growth.plot(kind='line', marker='o', figsize=(12,6))
plt.title("USA vs International Content Added Over Time on Netflix")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

# 19. Use pivot tables to analyze the relationship between rating (TV-MA, PG, etc.) and genre.

```python
df_pivot = df.copy()

df_pivot['Genre'] = df_pivot['listed_in'].str.split(', ')
df_pivot = df_pivot.explode('Genre')
```

```python
pivot_rating_genre = pd.pivot_table(
    df_pivot,
    values='title',
    index='Genre',
    columns='rating',
    aggfunc='count',
    fill_value=0
)

pivot_rating_genre.head()
```

```python
major_ratings = ['TV-MA', 'TV-14', 'PG', 'R', 'TV-PG']
pivot_rating_genre[major_ratings].sort_values(by='TV-MA', ascending=False).head(10)
```

```python
pivot_percent = pivot_rating_genre.div(
    pivot_rating_genre.sum(axis=1), axis=0
) * 100

pivot_percent[major_ratings].round(1).head(10)
```

```python
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(
    pivot_percent[major_ratings],
    cmap='coolwarm',
    annot=False
)

plt.title("Parental Rating Distribution Across Genres (%)")
plt.xlabel("Rating")
plt.ylabel("Genre")
plt.show()
```

# 20. Find the top 10 fastest-growing genres (year-over-year).

```python
# A genre is fast-growing if the number of titles added each year is increasing rapidly over time.
```

```python
df['YearAdded'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
```

```python
df_genre = df.copy()
df_genre['Genre'] = df_genre['listed_in'].str.split(', ')
df_genre = df_genre.explode('Genre')
```

```python
genre_year = (
    df_genre
    .groupby(['YearAdded', 'Genre'])
    .size()
    .reset_index(name='Count')
)
```

```python
genre_year['YoY_Growth'] = (
    genre_year
    .groupby('Genre')['Count']
    .diff()
)
```

```python
avg_growth = (
    genre_year
    .groupby('Genre')['YoY_Growth']
    .mean()
    .sort_values(ascending=False)
)
```

```python
top_10_fastest = avg_growth.head(10)
print(top_10_fastest)
```

```python
top_10_fastest.plot(
    kind='bar',
    figsize=(10,6),
)
plt.title("Top 10 Fastest-Growing Netflix Genres (YoY)")
plt.xlabel("Genre")
plt.ylabel("Average Year-over-Year Growth")
plt.xticks(rotation=45)
plt.show()
```
